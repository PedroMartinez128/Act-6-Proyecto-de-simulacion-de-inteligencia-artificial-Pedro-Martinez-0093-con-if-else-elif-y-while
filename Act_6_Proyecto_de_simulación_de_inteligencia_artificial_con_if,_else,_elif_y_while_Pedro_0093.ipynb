#### Problema No. 38 (Divercificacion de cartera).

```python
# Corrida 1: La prueba de alertas (Validaciones de fondos)
# --- SIMULADOR DE IA: DIVERSIFICACIÓN DE CARTERA ---

# 1. Configuración inicial y entrada de datos
print("--- BIENVENIDO A TU ASESOR FINANCIERO IA ---")
capital_inicial = float(input("Ingrese el capital total que desea invertir: ")) # [6, 7]
capital_disponible = capital_inicial # Variable para el seguimiento del saldo
inv_acciones = 0.0  # Acumulador para acciones
inv_bonos = 0.0     # Acumulador para bonos
inv_cripto = 0.0    # Acumulador para criptomonedas
continuar = "si"    # Variable de control del ciclo [2, 8]

# 2. Ciclo de interacción principal
# Se repite mientras el usuario quiera y tenga dinero disponible [1, 9]
while continuar == "si" and capital_disponible > 0: # [10, 11]
    print("\n--- MENÚ DE INVERSIÓN ---")
    print(f"Saldo disponible para invertir: ${capital_disponible}")
    print("1. Acciones (Riesgo Moderado)")
    print("2. Bonos (Riesgo Bajo)")
    print("3. Criptomonedas (Riesgo Alto)")

    opcion = int(input("Seleccione el número del activo (1-3): ")) # [4, 7]
    monto = float(input(f"¿Cuánto desea invertir en este activo? "))

    # Validación de fondos con operadores relacionales [11, 12]
    if monto > capital_disponible:
        print("IA Alerta: Fondos insuficientes. Intente un monto menor.")
    elif monto <= 0: # [4, 5]
        print("IA Alerta: El monto debe ser un valor positivo.")
    else:
        # Procesamiento de la inversión según la opción [5, 13]
        if opcion == 1:
            inv_acciones = inv_acciones + monto # Operador aritmético de suma [14, 15]
            print("IA Mensaje: Se han añadido Acciones a su cartera.")
        elif opcion == 2:
            inv_bonos = inv_bonos + monto
            print("IA Mensaje: Se han añadido Bonos (estabilidad) a su cartera.")
        elif opcion == 3:
            inv_cripto = inv_cripto + monto
            print("IA Mensaje: Se ha añadido Cripto (volatilidad) a su cartera.")
        else:
            print("IA Error: Opción de activo no reconocida.")
            # Si la opción no es válida, no restamos el capital
            capital_disponible = capital_disponible + monto

        # Actualización del capital restante [15]
        capital_disponible = capital_disponible - monto

    # Control del flujo interactivo [2, 16]
    if capital_disponible > 0:
        continuar = input("\n¿Desea realizar otra inversión? (si/no): ")
    else:
        print("\nTodo el capital ha sido asignado.")

# 3. Análisis lógico de la IA (Diagnóstico final)
print("\n--- ANÁLISIS DE LA IA SOBRE SU DIVERSIFICACIÓN ---")

# Cálculo de porcentajes usando operadores aritméticos [17, 18]
porcentaje_cripto = (inv_cripto / capital_inicial) * 100
porcentaje_bonos = (inv_bonos / capital_inicial) * 100

# Lógica de decisión de la IA basada en reglas condicionales [4, 5]
if porcentaje_cripto > 50: # [11, 19]
    print("DICTAMEN IA: CARTERA DE ALTO RIESGO. Su exposición a cripto es muy elevada.")
elif porcentaje_bonos > 70:
    print("DICTAMEN IA: CARTERA MUY CONSERVADORA. Demasiado peso en bonos, rendimiento potencial bajo.")
elif inv_acciones > 0 and inv_bonos > 0 and inv_cripto > 0: # Operador lógico 'and' [10, 11]
    print("DICTAMEN IA: EXCELENTE DIVERSIFICACIÓN. Ha cubierto todos los sectores de riesgo.")
else:
    print("DICTAMEN IA: CARTERA ACEPTABLE. Considere añadir activos en los sectores faltantes.")

print(f"\nCapital final invertido: ${capital_inicial - capital_disponible}")
print(f"Capital sobrante: ${capital_disponible}")
print("--- Fin de la asesoría ---")
print("programa realizado por Pedro Martinez NL = 38")
```
